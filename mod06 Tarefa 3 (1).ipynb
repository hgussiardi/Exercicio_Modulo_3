```python
#Fazendos os imports de bibliotecas e carregando arquivo
import sys
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_nasc = pd.read_csv('SINASC_RO_2019.csv')
df_nasc

#df_nasc['APGAR5']
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
#preparando variável APGAR5
#verificando se temos conteúdo sem preenchimento
apgar5_conteudo = df_nasc['APGAR5'].isna()
print('Resultado')
print(apgar5_conteudo.value_counts())

df_nasc['APGAR5'].dropna()
print(df_nasc['APGAR5'].value_counts())
print('Quantidade de Registro após limpeza')
print(df_nasc['APGAR5'].value_counts().sum())



cat_apgar5=[]
cat_apgar5=df_nasc['APGAR5'].map({0:'asfixia grave',1:'asfixia grave',2:'asfixia grave',3:'asfixia grave',
                4:'asfixia moderada',5:'asfixia moderada',6:'asfixia leve',7:'asfixia leve',
                8:'normal', 9:'normal', 10:'normal'}) 
df_nasc['Cat_APGAR5']=cat_apgar5
print(df_nasc['Cat_APGAR5'].value_counts())
print(df_nasc['GESTACAO'])
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
#df_nasc['APGAR5'].groupby('SEXO','SEMANAS')
df_nasc_masc = df_nasc[df_nasc['SEXO']== 'Masculino']
df_nasc_fem = df_nasc[df_nasc['SEXO']== 'Feminino']
print(len(df_nasc_masc))
print(len(df_nasc_fem))

plt.close('all')
plt.title('Masculino')
sns.countplot(x='APGAR5', hue='GESTACAO', data= df_nasc_masc)
plt.legend(loc = 2)
plt.show()

plt.close('all')
plt.title('Feminino')
sns.countplot(x='APGAR5', hue='GESTACAO', data= df_nasc_fem)
plt.legend(loc = 2)
plt.show()
#
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
tab_cruzada = pd.crosstab(df_nasc['GESTACAO'],df_nasc['APGAR5'])
tab_cruzada
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
tab_cruzada_pctlin = tab_cruzada.div(tab_cruzada.sum(axis=1), axis=0)
tab_cruzada_pctlin
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
plt.rc('figure', figsize=(18,8))
tab_cruzada_pctlin.plot.bar()
plt.legend(loc = 2)
```

Os dados do gráfio acima não corroboram com a classificação dos médicos, pois nascidos entre 32 e 36 semanas tem uma distribuição muito parecida com nascidos entre 37 e 41 semanas.
