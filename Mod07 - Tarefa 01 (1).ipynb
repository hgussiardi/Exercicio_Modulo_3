#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
df_demo_new.sexo = df_demo_new.sexo.map({'M':1,'F':0})
df_demo_new.posse_de_veiculo = df_demo_new.posse_de_veiculo.map({'Y':1,'N':0})
df_demo_new.posse_de_imovel = df_demo_new.posse_de_imovel.map({'Y':1,'N':0})
df_demo_new.mau = df_demo_new.mau.astype(int)

print(df_demo_new.info())

````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
tab_cruzada = pd.crosstab(df_demo_new['possui_email'],df_demo_new['mau'])
tab_cruzada_pctlin = tab_cruzada.div(tab_cruzada.sum(axis=1), axis=0)
tab_cruzada_pctlin
print(tab_cruzada_pctlin)

tab_cruzada = pd.crosstab(df_demo_new['posse_de_veiculo'],df_demo_new['mau'])
tab_cruzada_pctlin = tab_cruzada.div(tab_cruzada.sum(axis=1), axis=0)
print(tab_cruzada_pctlin)


````

Aparentemente as 2 váriáveis são muito similares, o percentual de pessoas que são mau pagadores e bom pagadores é muito similar independentemente se a pessoa possui ou não celular e possui ou não veiculo, em torno 2% para todos os casos.
